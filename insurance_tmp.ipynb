```python
# Inicialização do OneHotEncoder
onehot = OneHotEncoder(drop='if_binary', handle_unknown='ignore', sparse_output=False)
categorical_columns = ['sex', 'smoker', 'region']

# Ajuste do codificador apenas às colunas categóricas
onehot.fit(X_train[categorical_columns])

# Transforme os dados diretamente em um array NumPy (já que sparse_output=False)
X_train_transformed = onehot.transform(X_train[categorical_columns])

# Criar DataFrame do array NumPy
X_train_transformed = pd.DataFrame(X_train_transformed,
                                   columns=onehot.get_feature_names_out(categorical_columns),
                                   index=X_train.index)

# Juntar os novos dados transformados com as colunas restantes
X_train_final = X_train_transformed.join(X_train.drop(columns=categorical_columns))
print(X_train_final.head())

```

```python
onehot = OneHotEncoder(drop='if_binary', handle_unknown='ignore', sparse_output=False)

numeric_processor = make_pipeline(SimpleImputer(strategy='mean'), StandardScaler())
categorical_processor = make_pipeline(SimpleImputer(strategy='most_frequent'), onehot)

preprocessador = ColumnTransformer([
    ('onehot', categorical_processor, ['sex', 'smoker', 'region']),
    ('scale', numeric_processor, ['age', 'bmi', 'children'])
])

model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessador),
    ('lr_model', LinearRegression())
])

model_pipeline.fit(X_train, y_train)
```
